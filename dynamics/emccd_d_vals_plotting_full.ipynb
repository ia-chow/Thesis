```python
import numpy as np
import matplotlib.pyplot as plt
```

Import:

```python
# either dsh or dh depending on which one
log10_d_vals = np.log10(np.load('./emccd_d_vals/all_emccd_dh_vals.npy'))
```

Check to make sure it is the right length ($101631$ choose $2 = 5164379265$):

```python
log10_d_vals.shape
```

Plot in the same style as Fig. 11 of Shober et al. 2025:

```python
plt.figure(figsize = (12, 8))
# entries, edges, _ = plt.hist(log10_d_vals, bins=100, color='k', cumulative=True, histtype='step')
n_bins = 100
bins = np.linspace(np.min(log10_d_vals), np.max(log10_d_vals), n_bins)
entries, edges, _ = plt.hist(log10_d_vals, bins=bins, color='k', cumulative=True, histtype='step')
plt.yscale('log')
plt.xlim(-3, -0.35)
plt.xlabel(r'log $D_{H}$'), plt.ylabel(r'$N_{D_{H}} < $')
plt.savefig('emccd_dh_zoomed.png')
```
